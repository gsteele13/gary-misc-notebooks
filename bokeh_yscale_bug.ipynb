```python
import numpy as np

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook
from bokeh.models import LogScale, LinearScale


# This will import the latest compatible bokeh JS from the CDN
output_notebook()

# Generate data
x = np.linspace(0,100,1000)
y = 1/((x-50)**2 + 1) 
y += np.random.normal(size=len(x))*0.01
y += 1/((x-80)**2 + 5)

# Plot it
p = figure(height=200, width=600) 
p.sizing_mode = "scale_width"
p.line(x,y)
# If uncommented, this does work
# p.y_scale = LogScale()
target = show(p, notebook_handle=True)

# But this does nothing
p.y_scale = LogScale()
push_notebook(handle=target)
```

# Answer: not supported

https://github.com/bokeh/bokeh/issues/11983

Workaround is to make two plots and toggle the visibility. 

```python
import numpy as np

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook
from bokeh.models import LogScale, LinearScale

from bokeh.layouts import column
from ipywidgets import interact

p1 = figure(height=200, width=600) 
p1.sizing_mode = "scale_width"
p1.line(x,y)

p2 = figure(height=200, width=600)
p2.x_range = p1.x_range
p2.sizing_mode = "scale_width"
p2.line(x,y)
p2.y_scale = LogScale()
p2.visible = False

col = column([p1,p2])
col.sizing_mode = "scale_width"
target = show(col, notebook_handle=True)

def update(log=0):
    if log: 
        p1.visible = False
        p2.visible = True
    else:
        p1.visible = True
        p2.visible = False
    push_notebook(handle=target)

interact(update, log=(0,1))
```
